```python
# renaming columns
df = df.rename(columns = {"NO":"Not_Out","HS":"Highest Score","SR":"Strike Rate","100":"No_Of_Hundreds","50":"No_Of_Fifties","4s":"No_Of_Fours","6s":"No_Of_Sixies","BF":"Balls Faced"})
```

```python
# Checking Null Values
```

```python
df.isnull().any()
```

```python
# Fetching those rows where No_Of_Hundreds is NAN or null
df[df["No_Of_Hundreds"].isna()==True]
```

```python
# Fill Null with 0
df['No_Of_Hundreds'] = df['No_Of_Hundreds'].fillna(0)
df['No_Of_Fifties'] = df['No_Of_Fifties'].fillna(0)
df['No_Of_Fours'] = df['No_Of_Fours'].fillna(0)
df['No_Of_Sixies'] = df['No_Of_Sixies'].fillna(0)
```

```python
# If we want to check duplicates based on certain column, we use subset
# If want to keep the first duplicate value, then use keep='first' and for last we use keep='last'
# keep = 'first'
df2 =df.drop_duplicates(subset='Player',keep='first')
```

```python
df2[df2['Player'].isin(['K L Rahul','Hardik Pandya','Shubman Gill','Shikhar Dhawan','Sanju Samson'])]
```

```python
df3 =df.drop_duplicates(subset='Player',keep='last')
```
